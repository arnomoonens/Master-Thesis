# Activation functions

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Perceptron

```python
plt.plot(range(-10,1,1),[-1] * 11,color="#1f77b4")
plt.plot(range(0,11,1),[1] * 11, color="#1f77b4")
plt.plot([0,0],[-1,1],color="#1f77b4")
plt.xlabel("$x$")
plt.ylabel("output")
plt.ylim(-2,2)
plt.xlim(-10,10)
plt.axvline(x=0,color="black", linestyle="--")
```

## Sigmoid

```python
sigmoid_x = np.linspace(-10, 10, 100)
sigmoid_y = 1.0/(1+np.exp(-sigmoid_x))
```

```python
plt.plot(sigmoid_x,sigmoid_y,color="#1f77b4")
plt.xlabel("$x$")
plt.ylabel("output")
plt.xlim(-10,10)
plt.axvline(x=0,color="black", linestyle="--")
```

## Hyperbolic tangent (_tanh_)

```python
tanh_x = np.linspace(-10, 10, 100)
tanh_y = np.tanh(tanh_x)
```

```python
plt.plot(tanh_x,tanh_y)
plt.xlabel("$x$")
plt.ylabel("output")
plt.xlim(-10,10)
plt.axvline(x=0,color="black", linestyle="--")
```

## ReLU

```python
relu_x = np.concatenate((np.linspace(-10, 0, 101), np.linspace(0, 10, 101)),axis=0)
relu_y = [i if i >= 0 else 0 for i in relu_x]
```

```python
plt.plot(relu_x,relu_y)
plt.xlabel("$x$")
plt.ylabel("output")
plt.ylim(-0.5,2)
plt.xlim(-2,2)
plt.axvline(x=0,color="black", linestyle="--")
```

### Leaky ReLU

```python
alpha = 0.2
lrelu_x = np.concatenate((np.linspace(-10, 0, 101), np.linspace(0, 10, 101)),axis=0)
lrelu_y = [i if i >= 0 else alpha * i for i in lrelu_x]
```

```python
plt.plot(lrelu_x,lrelu_y)
plt.xlabel("$x$")
plt.ylabel("output")
plt.ylim(-1,2)
plt.xlim(-2,2)
plt.axvline(x=0,color="black", linestyle="--")
```
